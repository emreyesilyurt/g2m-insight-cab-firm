#### importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xlrd
from datetime import date
import warnings
import calendar
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
sns.set_palette("crest")
```

### loading datasets

```python
cab_data = pd.read_csv('Cab_Data.csv')
city = pd.read_csv('City.csv')
customer_id = pd.read_csv('Customer_ID.csv')
transaction = pd.read_csv('Transaction_ID.csv')
```

### Merging datasets

```python
temp = cab_data.merge(transaction, how = 'left', on = 'Transaction ID')
df = temp.merge(customer_id, how = 'left', on = 'Customer ID')
```

```python
df.columns
```

**Rename columns**

```python
df.rename(columns = {'Price Charged':'Price_Charged', 'Date of Travel':'Date_of_Travel' ,'KM Travelled':'KM_Travelled','Transaction ID':'Transaction_ID','Cost of Trip':'Cost_of_Trip', 'Income (USD/Month)':'Income_(USD/Month)'}, inplace = True)
```

### Grouping age column

```python
bins=[0,25,40,60,100]
labels = ['Young', 'Adult', 'Senior', 'Retired']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df.AgeGroup.unique()
```

### Add Profit column

```python
df['Profit'] = df['Price_Charged'] - df['Cost_of_Trip']
df['Profit_per_km'] = df['Profit'] / df['KM_Travelled']
```

### Edit datatime object

```python
df['Date_of_Travel'] = df['Date_of_Travel'].apply(lambda s: xlrd.xldate.xldate_as_datetime(s, 0))
```

```python
df['Year']= df['Date_of_Travel'].dt.year
df['Month']= df['Date_of_Travel'].dt.month
df['Day']= df['Date_of_Travel'].dt.day
```

```python
df.drop(columns = ['Date_of_Travel'], inplace=True)
```

### Add weekday

```python
temp_df = cab_data.copy()
temp_df['Date of Travel'] = temp_df['Date of Travel'].apply(lambda s: xlrd.xldate.xldate_as_datetime(s, 0))
df['Weekday'] = temp_df['Date of Travel'].apply(lambda s: calendar.day_name[s.weekday()])
```

### Outlier detection

```python
columns = ['KM_Travelled', 'Price_Charged', 'Cost_of_Trip', 'Age', 'Income_(USD/Month)']


for i in columns:
    plt.figure()
    sns.boxplot(x=df[i])
    plt.show()
```

```python
Q1 = df['Price_Charged'].quantile(0.25)
Q3 = df['Price_Charged'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_lim = Q1 - 1.4 * IQR
upper_lim = Q3 + 1.4 * IQR
```

```python
outliers_low = (df['Price_Charged'] < lower_lim)
outliers_up = (df['Price_Charged'] > upper_lim)
```

```python
len(df['Price_Charged'] - (len(df['Price_Charged'][outliers_low] + len(df['Price_Charged'][outliers_up]))))
df['Price_Charged'][(outliers_low | outliers_up)]
df['Price_Charged'][~(outliers_low | outliers_up)]
df = df[~(outliers_low | outliers_up)]
```

```python
plt.figure()
sns.boxplot(x=df['Price_Charged'])
plt.show()
```

### Correlation

```python
f, ax = plt.subplots(figsize=(12, 9))
corrmat = df.corr()
mask = np.triu(np.ones_like(corrmat, dtype=bool))

sns.heatmap(corrmat,mask=mask, cmap='crest',annot=True ,vmax=1,vmin=-1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}) 
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
